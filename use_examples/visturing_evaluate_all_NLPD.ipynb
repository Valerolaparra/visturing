# Evaluation of NLPD metric using visturing


This is an example to illustrate the use of visturing toolbox to evaluate a perceptual metric. In this case we evaluate the NLPD.

```python
from perceptual_quality import nlpd
```

```python

def evaluate_dist(ImA,ImB):
    # Imagenes dimensiones (Nims,H,W,C)
    # Normalizadas a [0,255]
    image_A = tf.convert_to_tensor(ImA)
    image_B = tf.convert_to_tensor(ImB)

    dists = np.zeros(ImA.shape[0])
    for i in range(ImA.shape[0]):
        distance = nlpd.nlpd(image_A[i], image_B[0])
        dists[i] = np.mean(distance.numpy())

    return dists
```

```python
# use example
import numpy as np
import tensorflow as tf
ImA = np.abs(255*np.random.rand(5,128,128,3))
ImB = np.abs(255*np.random.rand(1,128,128,3))

evaluate_dist(ImA,ImB)
```

# Genera Tabla

```python
def generate_results(data_path,gt_path):
    from visturing.properties import prop1, prop2, prop3_4, prop5, prop6_7, prop8, prop9, prop10

    out1 = prop1.evaluate(evaluate_dist,data_path+'/Experiment_1',gt_path)
    print('out1 done')
    out2 = prop2.evaluate(evaluate_dist,data_path+'/Experiment_2',gt_path)
    print('out2 done')
    out3_4 = prop3_4.evaluate(evaluate_dist,data_path+'/Experiment_3_4',gt_path)
    #out5 = prop5.evaluate(evaluate_dist,data_path+'/Experiment_5',gt_path)
    out6_7 = prop6_7.evaluate(evaluate_dist,data_path+'/Experiment_6_7',gt_path)
    print('out6_7 done')
    out8 = prop8.evaluate(evaluate_dist,data_path+'/Experiment_8',gt_path)
    print('out8 done')
    out9 = prop9.evaluate(evaluate_dist,data_path+'/Experiment_9',gt_path)
    print('out9 done')
    out10 = prop10.evaluate(evaluate_dist,data_path+'/Experiment_10',gt_path)
    print('out10 done')

    correlations =  [out1['corr'], out2['correlations'], out6_7['pearson'], out8['pearson']] # [out1['corr'], out2['correlations'], out3_4['pearson_corr'], out5['pearson_corr'], out6_7['pearson'], out8['pearson']]
    return correlations,out1, out2, out6_7, out8, out9, out10 #correlations,out1, out2, out3_4, out5, out6_7, out8, out9, out10
```
