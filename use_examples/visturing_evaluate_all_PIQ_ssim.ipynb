```python

import torch
from piq import DISTS, LPIPS, ssim

def evaluate_dist(ImA,ImB):
    # Imagenes dimensiones (Nims,H,W,C)
    # Normalizadas a [0,255]
    image_A = torch.Tensor(ImA/255).permute(0, 3, 1, 2)
    image_B = torch.Tensor(ImB/255).permute(0, 3, 1, 2)

    
    distances = np.zeros(ImA.shape[0])

    for i in range(ImA.shape[0]):
        distances[i]: torch.Tensor = ssim(image_A[i:i+1], image_B[0:1])
        #distances[i]: torch.Tensor = DISTS(image_A[i:i+1], image_B[0:1])
        #distances[i]: torch.Tensor = ssim(image_A[i:i+1], image_B[0:1], data_range=1.)

    return distances
    
```

```python
import numpy as np
# Testing the function behaviour
ImA = np.abs(255*np.random.rand(5,128,128,3))
ImB = np.abs(255*np.random.rand(1,128,128,3))

evaluate_dist(ImA,ImB)
```

```python
# This function evaluates all the experiments proposed in the paper
# The first time it downloads the data, so it will take long
# If the data is already downloaded it should take a couple of seconds

results = evaluate_all(evaluate_dist,
             data_path="./Data/",
             gt_path="./Data/")
```
